```python
# Labs 6

import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error


clean_data2
```

```python
clean_data2.dtypes
```

```python
# X-y split

y = clean_data2['TOTAL_CLAIM_AMOUNT']
X = clean_data2.drop(['TOTAL_CLAIM_AMOUNT','CUSTOMER'], axis=1)
```

```python
# Numerical / categorical partition

clean_num = X.select_dtypes(include = np.number)
clean_cat = X.select_dtypes(include = np.object)
```

```python
clean_num.dtypes
```

```python
clean_num
```

```python
clean_cat.dtypes
```

```python
clean_cat
```

```python
# I can do it with get_dummies or with the encoder! It is up to me!

#dm_table = pd.get_dummies(clean_cat)
#dm_table.shape
```

```python
# Encoding data

encoder = OneHotEncoder(handle_unknown='error', drop='first')
encoder.fit(clean_cat)
```

```python
encoded = encoder.transform(clean_cat).toarray()
encoded
```

```python
encoder.categories_
```

```python
X = np.concatenate([clean_num, encoded], axis=1)
X.shape
```

```python
# traint-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
r2_score(y_test, predictions), mean_squared_error(y_test, predictions, squared=False), mean_squared_error(y_test, predictions)
```
